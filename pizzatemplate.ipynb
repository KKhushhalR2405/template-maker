```python
from matplotlib import pyplot as plt
import cv2 as cv2
import pandas as pd
import PIL
from PIL import ImageFont
from PIL import Image
from PIL import ImageDraw
```

```python
file = pd.read_csv("pizza images.csv",sep=',')
```

```python
file
```

```python
img = Image.open("template 3.jpg")

print("Template")


im = img.copy()
im = im.resize((1080,1500))
plt.figure()
plt.imshow(im)
```

```python
i = Image.open("pizza 3.jpg")
i = i.resize((200,300))
plt.figure()
plt.imshow(i)
```

```python
l=[]
for j in range(8):
    i = Image.open(file["pizza image"][j])
    i = i.resize((200,300))
    
    
    img1 = ImageDraw.Draw(i)   
    shape = [(120,235), (250,350)]

    img1.ellipse(shape, fill=(255,255,0), outline=(255,255,0))

    font = ImageFont.truetype('Montserrat-Bold.ttf', 12)
    text = "       " + str(file['discount'][j] ) + "%" +" off !!" + "\n" + "Now @ " + str(file['price'][j]) + "/-"

    img1.text((125,250), text, (255,0,0), font=font)

    path = "outputtest" + str(j) + ".png"
    l.append(path)
    
    plt.figure()
    plt.imshow(i)
    
    i.save(path)







```

```python
d = {1:(61,69) , 2:(400,69) , 3:(740,69) , 4:(61,550) , 5:(740,550) , 6:(61,1031) , 7:(400,1031) , 8:(740,1031)}
imgg=img.copy()
```

```python


for k in range(8):
    
    final = Image.open(l[k])
    final = final.resize((270,370))
    imgg.paste(final,d[k+1])
    plt.figure()
    plt.imshow(imgg)
    
imgg.save("final ouput.png")
```
